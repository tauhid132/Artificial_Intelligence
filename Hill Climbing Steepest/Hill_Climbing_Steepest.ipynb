```python
def init():
  data_list = [2, 1, 5, 0, 8, 4, 10, 0, 20, 10]
  return data_list
```

```python
def calc_cost(state):
  disorders = 0
  for i in range(len(state)):
    for j in range(i, len(state)):
      if(state[j] < state[i]):
        disorders +=1
  return disorders
```

```python
def generate_state(currentState, currentStateCost):
  minCost = 9999
  for i in range(len(currentState)-1):
    for j in range(i+1,len(currentState)):
      newGenState = currentState.copy()
      temp = newGenState[i]
      newGenState[i] = newGenState[j]
      newGenState[j] = temp
      newStateCost = calc_cost(newGenState)
      if newStateCost < minCost:
        minCost = newStateCost
        newState = newGenState.copy()
  if(minCost < currentStateCost):
    return newState, minCost
  else:
    return currentState, None
      
```

```python
def goal_test(state):
  if calc_cost(state) == 0:
    return True
  else:
    return False
```

```python
state = init()
while not goal_test(state):
  cost = calc_cost(state)
  state, cost = generate_state(state, cost)
  if(cost == None):
    print(state)
    break;
  print(state)


```
