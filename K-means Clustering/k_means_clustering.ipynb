```python
from numpy import genfromtxt
data_path = '/content/gdrive/MyDrive/AI_LAB/Assignment-3/data.csv'
my_data = genfromtxt(data_path, delimiter=',')
```

```python
import random
import numpy as np
centers = []
np.random.shuffle(my_data)
centers = my_data[:6]
print(centers)
clusters = [[],[],[],[],[],[]]
```

```python
iteration = 0
while(1):
  temp_clusters = [[],[],[],[],[],[]]
  for s in range(len(my_data)):
    distance = []
    for i in range(6):
      dist = np.linalg.norm(my_data[s] - centers[i])
      distance.append(dist)
    min_center_index = np.argmin(distance) 
    temp_clusters[min_center_index].append(s) 
  index = 0
  
  for L in temp_clusters:
    total_x = 0
    total_y = 0
    for E in L: 
      total_x = total_x + my_data[E][0] 
      total_y = total_y + my_data[E][1]
    x = total_x / len(L)
    y = total_y / len(L)
    centers[index] = [x,y]
    index+=1
  
  iteration+=1
  if iteration > 1:
    shift = 0
    for s in range(len(my_data)):
      location_in_clusters = -1
      for c in range(len(clusters)):
        for d in range(len(clusters[c])):
          if clusters[c][d] == s:
            location_in_clusters = c
            break;
      
      location_in_temp_clusters = -1
      for t in range(len(temp_clusters)):
        for p in range(len(temp_clusters[t])):  
          if temp_clusters[t][p] == s: 
            location_in_temp_clusters = t
            break;
      
      if location_in_clusters != location_in_temp_clusters:
        shift+=1
    shift_percent = shift / len(my_data) *100
    if(shift_percent < 0.1):
      break;    
  clusters = temp_clusters.copy()
```

```python
import matplotlib.pyplot as plt

list_1_x = []
list_1_y = []
for i in clusters[0]:
  list_1_x.append(my_data[i][0])
  list_1_y.append(my_data[i][1])

list_2_x = []
list_2_y = []
for i in clusters[1]:
  list_2_x.append(my_data[i][0])
  list_2_y.append(my_data[i][1])

list_3_x = []
list_3_y = []
for i in clusters[2]:
  list_3_x.append(my_data[i][0])
  list_3_y.append(my_data[i][1])

list_4_x = []
list_4_y = []
for i in clusters[3]:
  list_4_x.append(my_data[i][0])
  list_4_y.append(my_data[i][1])

list_5_x = []
list_5_y = []
for i in clusters[4]:
  list_5_x.append(my_data[i][0])
  list_5_y.append(my_data[i][1])

list_6_x = []
list_6_y = []
for i in clusters[5]:
  list_6_x.append(my_data[i][0])
  list_6_y.append(my_data[i][1])


plt.scatter(list_1_x, list_1_y, color="green") 
plt.scatter(list_2_x, list_2_y, color="blue") 
plt.scatter(list_3_x, list_3_y, color="red")
plt.scatter(list_4_x, list_4_y, color="orange") 
plt.scatter(list_5_x, list_5_y, color="pink") 
plt.scatter(list_6_x, list_6_y, color="grey")
plt.scatter(centers[:,0], centers[:,1], color="black")
```
